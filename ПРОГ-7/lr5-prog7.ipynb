```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
html = urlopen('https://www.herzen.spb.ru/main/structure/inst/')
bs = BeautifulSoup(html, "html.parser")
# nameList = bs.find('td',{'class':'block'}).children
```

```python
tdList = bs.find('td', {'class': 'blockwh'})

instRawList= tdList.div.ul.contents
print(instRawList)
instList = []
for el in instRawList:
  instList.append(el.a.get_text())
print(instList)
linksList = []
for el in instRawList:
  linksList.append(el.a['href'])
print(linksList)

futureJson = []
```

```python

# парсинг атласа
atlas_html = urlopen('https://atlas.herzen.spb.ru/faculty.php')
bs2 = BeautifulSoup(atlas_html, "html.parser")
atlasList = bs2.findAll('li',{'fac':1})
print(atlasList)

atlasInstList = []
for el in atlasList:
  atlasInstList.append(el.a.get_text())
print(atlasInstList)
```

```python
import re
for el in instRawList:
  instName = el.a.get_text() # получаем названия институтов
  instNameEdit = instName[7:len(instName)-1:] #созадем шаблон по которому будем сопостовлять названия институтов с разных страниц
  #print(instNameEdit)
  instLink = el.a['href'] # получаем ссылки на институты
  #print(instLink)
  x = bs2.find("a",string=re.compile(instNameEdit, re.IGNORECASE)) # Ищем элемент с тэгом a, в котором нужное нам название института
  #print(x)
  par_x = x.find_parent('li') # Ищем родителя найденного элемента
  kafList= par_x.ul.findAll('a', string=re.compile("кафедра")) # Ищем элементы с названием кафедр
  kafedrasList = []
  for el in kafList:
    urlPart = str(el['href']) # получаем относительные ссылки на информацию о кафедрах
    url = "https://atlas.herzen.spb.ru/"+urlPart # создаем абсолютные ссылки
    kafHtml = urlopen(url) # переходим по ссылке
    kafBs = BeautifulSoup(kafHtml, "html.parser") # парсим
    zavKaf = kafBs.find("h3",string=re.compile("Заведующий")) # ищем заголовок завкафедры
    zavKafName = zavKaf.find_next_sibling().find("span").get_text().split(',')[0] #ищем имя завкафедры, возвращаем текст и отделяем часть с регалиями
    kafInfo = {"dep_name":el.get_text(),"head_name":zavKafName} # собираем инфу о кафедре в словарь
    kafedrasList.append(kafInfo) #добавляем кафедру в список
  futureJson.append({"institute_name":instName,"url":instLink,"kafedras":kafedrasList}) #собираем все в один объект

print(futureJson)
```
