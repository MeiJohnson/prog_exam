### Eventloop (Событийный цикл)

__Файловый дескриптор__ - число, с которым ассоциируется запущенный файл в системе.
__Select__ мониторит измененние тех файловых объектов, которые мы в нее передали. Получате на вход 3 списка:
+ Объекты, за которыми нужно наблюдать, когда они станут доступны для чтения.
+ Объекты, за которыми нужно наблюдать, когда они станут доступны для записи.
+ Объекты, у которых мы ожидаем какие-то ошибки.
Возвращает списки объектов, когда они станут доступны.

Модуль selectors обеспечивает высокоуровневое и эффективное мультиплексирование ввода-вывода, основанное на примитивах модуля select. Если пользователям не нужен точный контроль над примитивами операционной системы, то рекомендуется использовать этот модуль вместо модуля select.
